```python
from pydantic import BaseModel
from motor.motor_asyncio import AsyncIOMotorClient

# MongoDB connection URL
client = AsyncIOMotorClient(MONGO_URL)
database = client["turkgpt"]

modelsDB = database["models"]

class Model(BaseModel):
    name: str
    pages: list

async def create_new_model(model: Model):
    await modelsDB.insert_one(model.dict())
    return model

async def update_model(model: Model):
    await modelsDB.update_one({"name": model.name}, {"$set": model.dict()})
    return model

async def append_to_model_pages(model_name: str, new_pages: list):
    model = await modelsDB.find_one({"name": model_name})
    if model:
        current_pages = model.get("pages", [])
        updated_pages = current_pages + new_pages
        await modelsDB.update_one({"name": model_name}, {"$set": {"pages": updated_pages}})
        return {"name": model_name, "pages": updated_pages}
    else:
        raise ValueError(f"Model with name {model_name} not found")

async def get_all_models():
    models = []
    async for model in modelsDB.find({}):
        models.append(model)
    return models
```

```python
import asyncio

loop = asyncio.get_running_loop()

# Function to retrieve and print all models
async def fetch_and_print_models():
    models = await get_all_models()
    return models

task = loop.create_task(fetch_and_print_models())
```

```python
from unsloth import FastLanguageModel
from transformers import AutoTokenizer,AutoModelForCausalLM
import torch

max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "turkgpt/turkgpt-v0.1",
    max_seq_length = max_seq_length,
    dtype = dtype,
)

FastLanguageModel.for_inference(model) # Enable native 2x faster inference
```

```python
def format_prompt(message, history):
  liste = []
  liste.append({"role": "system", "content": "You are Turk-GPT, a helpful assistant and you are desiigned by Turk-GPT Team."})
  for i in history:
    liste.append({"role": "user", "content": i[0]})
    liste.append({"role": "assistant", "content": i[1]})
  message = {"role": "user", "content": message}
  liste.append(message)
  prompt = tokenizer.apply_chat_template(liste, tokenize=False, add_generation_prompt=True)
  return prompt
```

```python
from langchain_core.documents import Document

def get_document_pages(pages):
  return [Document(page_content=page["page_content"],metadata=page["metadata"]) for page in pages]
```

```python
g_retriever = None
g_rag_name = ""
```

```python
def get_rag_answer(prompt,rag_name):
    global g_retriever
    global g_rag_name

    retriever = None

    if g_rag_name == rag_name:
        retriever = g_retriever
    else:
      rag_model = {}

      for i in range(len(models_names)):
          if models_names[i] == rag_name:
              rag_model = models[i]


      # Load chunked documents into the FAISS index
      db = FAISS.from_documents(
          get_document_pages(rag_model["pages"]),
          HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2')
      )

      # Connect query to FAISS index using a retriever
      retriever = db.as_retriever(
          search_kwargs={'k': 4},
          search_type="similarity"
      )
      g_retriever = retriever
      g_rag_name = rag_name


    prompt_template = """
        Aşağıdaki metine bakarak soruyu cevapla:
        {context}
        Soru:
        {question}
    """

    # Create prompt from prompt template
    prompt_temp = PromptTemplate(
        input_variables=["context", "question"],
        template=prompt_template,
    )

    return prompt_temp.format(context=retriever.invoke(prompt), question=prompt)
```

```python
import gradio as gr
def generate(
    prompt, history, max_new_tokens=128,rag_model="default"
):
    try:
        if rag_model != "default":
            prompt = get_rag_answer(prompt,rag_model)

        formatted_prompt = format_prompt(prompt, history)


        inputs = tokenizer([formatted_prompt], return_tensors="pt").to("cuda")

        generated_text = model.generate(**inputs, max_new_tokens=max_new_tokens)
        output = tokenizer.decode(generated_text[0], skip_special_tokens=True)

        answer = output.split("assistant\n")[-1]

        return answer
    except Exception as e:
        return f"Error: {str(e)} {str(history)}"

additional_inputs=[
    gr.Slider(
        label="Max new tokens",
        value=128,
        minimum=0,
        maximum=1048,
        step=64,
        interactive=True,
        info="The maximum numbers of new tokens",
    ),
    gr.Dropdown(
        label="Selecet Rag Model",
        choices=rag_models,
        value="default",
    )
]

logo_html = """
<div style="display: flex;align-items: center;margin-bottom: 20px;justify-content: center;text-align: center;">
<a href="https://turk-gpt.com" style="text-decoration: none; color: inherit; display: flex; align-items: center;">
<img style="height: 50px;" alt="Logo" src="https://turk-gpt.com/static/media/logo-light.a4b1391c71003f720bf6.webp"> Turk-GPT
</a>
</div>"""

gr.ChatInterface(
    fn=generate,
    chatbot=gr.Chatbot(show_label=False, show_share_button=False, show_copy_button=True, likeable=True, layout="panel"),
    additional_inputs=additional_inputs,
    title=logo_html
).launch(show_api=False)
```
